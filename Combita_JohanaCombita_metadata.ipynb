Ahora vamos a ver cómo podemos manejar estos datos en Python. Para ello, utilizaremos la librería xml.

Para crear un documento XML bien formado, es necesario definir dónde está descrito el prefijo Dublin Core o "dc:". Para ello, añadimos antes de los datos la siguiente cabecera:

```XML
<?xml version="1.0" encoding="UTF-8" standalone="no"?><?xml-stylesheet type="text/xsl" href="/webservices/catalog/xsl/searchRetrieveResponse.xsl"?>
<searchRetrieveResponse xmlns:oclcterms="http://purl.org/oclc/terms/" xmlns:dc="http://purl.org/dc/elements/1.1/" xmlns:diag="http://www.loc.gov/zing/srw/diagnostic/" xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance">
```

Sin olvidar añadir al final:

```XML
</searchRetrieveResponse>
```

````python
import xml.etree.ElementTree as ET
dc_xml = '''<?xml version="1.0" encoding="UTF-8" standalone="no"?><?xml-stylesheet type="text/xsl" href="/webservices/catalog/xsl/searchRetrieveResponse.xsl"?>
<searchRetrieveResponse xmlns:oclcterms="http://purl.org/oclc/terms/" xmlns:dc="http://purl.org/dc/elements/1.1/" xmlns:diag="http://www.loc.gov/zing/srw/diagnostic/" xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance">


     <dc:contributor>Edvard Munch </dc:contributor>
  <dc:coverage>Lugar indeterminado</dc:coverage>
  <dc:creator>Edvard Munch </dc:creator>
  <dc:date>1910</dc:date>
  <dc:description>Cuadro...</dc:description>
  <dc:format>Oleo sobre carton</dc:format>
  <dc:identifier>id_museo_grito</dc:identifier>
  <dc:language></dc:language>
  <dc:publisher>Galeria nacional de Oslo</dc:publisher>
  <dc:relation>cuadro1, cuadro2, cuadro3</dc:relation>
  <dc:rights>Acceso al museo</dc:rights>
  <dc:source></dc:source>
  <dc:title>El grito</dc:title>
  <dc:type>Cuadro</dc:type>



</searchRetrieveResponse>'''

tree = ET.fromstring(dc_xml)
tree
````

Si queremos recorrer los elementos del XML que hemos formado, podemos utilizar un bucle, teniendo en cuenta que la información que nos interesa la tenemos en el elemento raíz 'searchRetrieveResponse':

````python
for table in tree.iter('searchRetrieveResponse'):
    for child in table:
        print(child.tag, child.text)
````

Observa que, al utilizar el prefijo 'dc:' e indicarle que está descrito en la URL 'http://purl.org/dc/elements/1.1/', la eqtiqueta o "tag" aparece como, por ejemplo {URL}contributor.

Utilizando findall() sobre el arbol (tree), podemos encontrar todos los elementos con una etiqueta determinada.

````python
relation = tree.findall('{http://purl.org/dc/elements/1.1/}relation')
print(relation)
````

Ten en cuenta que lo que encontramos es, en realidad, una parte del documento XML, por lo que hay que iterarlo como antes:

````python
for child in relation:
    print(child.tag, child.text)
````

XML utiliza prefijos para no necesitar referenciar a la URL de un tipo cada vez, lo podemos ver en la cabecera:

```XML
<?xml version="1.0" encoding="UTF-8" standalone="no"?><?xml-stylesheet type="text/xsl" href="/webservices/catalog/xsl/searchRetrieveResponse.xsl"?>
<searchRetrieveResponse xmlns:oclcterms="http://purl.org/oclc/terms/" xmlns:dc="http://purl.org/dc/elements/1.1/" xmlns:diag="http://www.loc.gov/zing/srw/diagnostic/" xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance">'''
```

Por ejemplo, cada vez que queremos utilizar un tipo de Dublin Core, utilizamos el prefijo dc: que equivale a llamar a la definición:

xmlns:dc="http://purl.org/dc/elements/1.1/"

Sin embargo, para utilizar ElementTree en Python, tenemos que utilizar la URL completa. Esto puede resultar un poco engorroso, así que podemos definir el namespace para utilizar también el prefijo:

````python
namespaces = {'dc': 'http://purl.org/dc/elements/1.1/'} # add more as needed

tree.find('dc:rights',namespaces).text
````

````python
# Importo paquetes a utilizar 
import xml.etree.ElementTree as ET # trabajar XML en python
import requests # trabajar con servicios web

#obteniendo datos y guardandolos
response_dc = requests.get('https://schema.datacite.org/meta/kernel-4.5/example/datacite-example-full-v4.xml')
if response_dc.status_code == 200:
    with open("./datacite-example-full-v4.xml", 'wb') as f:
        f.write(response_dc.content)
        print(response_dc.status_code)
else:
    print(f"Error al obtener el archivo XML: {response_dc.status_code}")       
````

````python
#Se carga en python y vamos a la raiz 
tree_dc = ET.parse('datacite-example-full-v4.xml')
root_dc = tree_dc.getroot()

#Identificando nodos raiz:
for table in root_dc.iter():
    for child in table:
        if len(child)==0:
            print(child.tag, child.text)
````

````python
#Datos a mapaer en xml (Solución final)
dic_dublin_core = {"Title": ".//{http://datacite.org/schema/kernel-4}title",
    "Creator": ".//{http://datacite.org/schema/kernel-4}creatorName",
    "Subject": ".//{http://datacite.org/schema/kernel-4}subject",
    "Description": ".//{http://datacite.org/schema/kernel-4}description",
    "Publisher": ".//{http://datacite.org/schema/kernel-4}publisher",
    "Contributor": ".//{http://datacite.org/schema/kernel-4}contributorName",
    "Date": ".//{http://datacite.org/schema/kernel-4}publicationYear",
    "Type": ".//{http://datacite.org/schema/kernel-4}resourceType",
    "Format": ".//{http://datacite.org/schema/kernel-4}format",
    "Identifier": ".//{http://datacite.org/schema/kernel-4}identifier",
    "Source": ".//{http://datacite.org/schema/kernel-4}relatedItemIdentifier",
    "Language": ".//{http://datacite.org/schema/kernel-4}language",
    "Relation": ".//{http://datacite.org/schema/kernel-4}relatedIdentifier",
    "Coverage": ".//{http://datacite.org/schema/kernel-4}geoLocationPlace",
    "Rights": ".//{http://datacite.org/schema/kernel-4}rights"}

#diccionario de elementos vacio
elementos_dc = {}

# Extraer y mostrar los términos de Dublin Core
for dc_key, xpath in dic_dublin_core.items():
    elementos = root_dc.findall(xpath)
    # Inicialización de una lista para cada clave en el diccionario
    if dc_key not in elementos_dc:
        elementos_dc[dc_key] = []
        # Agregar los textos de los elementos encontrados
        for e in elementos:
            if e.text:  # Asegurar que el elemento tiene texto
                elementos_dc[dc_key].append(e.text) # agregar elementos a diccionario


# Obtener los elementos del diccionario
for key, values in elementos_dc.items():
    if values:  # Solo imprime si hay valores asociados a la clave
        print(f"- {key}: {', '.join(values)}")
    else:
        print(f"- {key}: No hay datos disponibles")
````

## Ejercicio 2

Haz un listado de todas las etiquetas del documento XML con sus atributos (si lo tienen)

````python
#identificando etiquetas de elementos con sus atributos (exploración de atributos)
for elemento in root_dc.iter():
    if elemento.attrib: 
        print(f'Etiqueta: {elemento.tag} | Atributos: {elemento.attrib}')
````

````python
#Solución final: Se eliminan los namespaces y se recorre el xml para identificar atributos.
for elemento in root_dc.iter():
    if '}' in elemento.tag: # Se eliminan los namespace de las etiquetas que tengan
        etiqueta = elemento.tag.split('}')[-1]
    else:
        etiqueta = elemento.tag

    # Revisión de atributos para salida por elemento
    if elemento.attrib:  
        atributo = {}  # Crear un diccionario vacío para almacenar los atributos
        for k, v in elemento.attrib.items():
            if '}' in k:
                # Si elimina namespace del atributo (si aplica)
                clave_atributo = k.split('}')[-1]
            else:
                clave_atributo = k
            # Agregar el atributo al diccionario
            atributo[clave_atributo] = v
        
        # Mostrar resultados de etiquetas y atributos del elemento
        print(f"- Etiqueta: {etiqueta} | Atributos: {atributo}")
````

````python
#solución final
#creando namespace
namespace = {'eml': 'eml://ecoinformatics.org/eml-2.1.1'}

# Buscar los elementos atributo
list_at = root_amt.findall('.//attributeList/attribute', namespace)

# Diccionario para almacenar los resultados
attributes_dict = {}

#Consultando elemento atributo para extraer el nombre y su unidad
#la cual se puede enocntrar de dos maneras: standardUnit o customUnit
for atr in list_at:
    # Obtener el nombre del atributo
    at_name_elem = atr.find('attributeName', namespace)
    at_name = at_name_elem.text if at_name_elem is not None else "Sin nombre"

    # Obtener las unidades
    std_unit = atr.find('.//standardUnit', namespace)
    c_unit = atr.find('.//customUnit', namespace)

    #buscando unidades y guardandolas
    if std_unit is not None:
        unit = std_unit.text
    elif c_unit is not None:
        unit = c_unit.text
    else:
        # Crear un sub-elemento customUnit si no existe ninguna unidad
        unit_elem = ET.SubElement(atr, "customUnit")
        unit_elem.text = "Sin unidad"
        unit = "Sin unidad"

    # Guardar en el diccionario
    attributes_dict[at_name] = unit

# Guardar el archivo modificado
tree_amt.write('./amt_prototype_with_units.xml')

# Consultar el diccionario y mostrar los resultados
for name, unit in attributes_dict.items():
    print(f"Nombre atributo: {name} | Unidad: {unit}")
````
